```python
member_length = data[['Length of Membership']]
member_length['Length of Membership'].fillna(data['Length of Membership'].mean(), inplace = True)
amount_spent = data[['Yearly Amount Spent']]
amount_spent['Yearly Amount Spent'].fillna(data['Yearly Amount Spent'].mean(), inplace = True)
amount_spent.head(40)
```

```python
plt.scatter(x = member_length['Length of Membership'], y = amount_spent['Yearly Amount Spent'], c = 'r')
```

```python
data['Yearly Amount Spent'].mean()
```

```python
#correlation when using mean
print(member_length['Length of Membership'].corr(amount_spent['Yearly Amount Spent']))
#strong correlation and better than the correlation when using zeros
```

```python
member_length = data[['Length of Membership']]
member_length['Length of Membership'].fillna(data['Length of Membership'].median(), inplace = True)
amount_spent = data[['Yearly Amount Spent']]
amount_spent['Yearly Amount Spent'].fillna(data['Yearly Amount Spent'].median(), inplace = True)
amount_spent.head(40)
```

```python
plt.scatter(x = member_length['Length of Membership'], y = amount_spent['Yearly Amount Spent'], c = 'r')
```

```python
#correlation when using median 
print(member_length['Length of Membership'].corr(amount_spent['Yearly Amount Spent']))
#strong correlation but not as good as when we used the mean
```

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
member_length = data[['Length of Membership']]
member_length['Length of Membership'].fillna(data['Length of Membership'].mean(), inplace = True)
amount_spent = data[['Yearly Amount Spent']]
amount_spent['Yearly Amount Spent'].fillna(data['Yearly Amount Spent'].mean(), inplace = True)
data['member_length_mean'] = member_length['Length of Membership'] 
data['amount_spent_mean'] = amount_spent['Yearly Amount Spent']
df = data[['amount_spent_mean', 'amount_spent_mean','Avg. Session Length','Time on App','Time on Website']]
scaler = StandardScaler()
scaler.fit(df)
```

```python
df
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```
